```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, t
import scipy.stats as stats


# Оценка матрицы ковариаций
def estimate_covariance_matrix(samples):
    # Вычисление выборочной матрицы ковариаций
    # rowvar=False означает, что каждая строка - это наблюдение, а столбец - переменная
    return np.cov(samples, rowvar=False)  

# Проверка гипотезы о коэффициенте корреляции
def test_correlation_hypothesis(samples, i, j, rho_0, alpha=0.05):
    # Выбор i-й и j-й переменных из выборки
    X_i = samples[:, i]
    X_j = samples[:, j]
    # Вычисление выборочного коэффициента корреляции Пирсона и p-value
    r, _ = pearsonr(X_i, X_j)
    N = len(X_i)
    
    # Вычисление t-статистики для проверки гипотезы
    t_stat = (r - rho_0) * np.sqrt(N - 2) / np.sqrt(1 - r**2)
    # Вычисление p-value с использованием t-распределения
    p_value = 2 * t.sf(np.abs(t_stat), df=N - 2)
    
    # Вывод результатов теста
    print(f"Выборочный коэффициент корреляции: {r:.4f}")
    print(f"p-value: {p_value:.4e}")
    
    # Принятие решения о гипотезе H0
    if p_value > alpha:
        print("Гипотеза H0 не отвергается.")
    else:
        print("Гипотеза H0 отвергается.")

# Параметры для многомерного нормального распределения
mean = np.array([0, 0])       # Вектор средних значение (математическое ожидание)
cov = np.array([[1, 0.5],     # Матрица ковариаций
                [0.5, 1]])
sample_size = 1000   # Размер выборки

# Генерация выборки
samples = np.random.multivariate_normal(mean, cov, sample_size)

# Оценка матрицы ковариаций
estimated_cov = estimate_covariance_matrix(samples)
print("Оцененная матрица ковариаций:")
print(estimated_cov)

# Проверка гипотезы о коэффициенте корреляции
i, j = 0, 1  # Индексы переменных для проверки корреляции
rho_0 = 0.5  # Гипотетическое значение коэффициента корреляции
test_correlation_hypothesis(samples, i, j, rho_0)
```

```python
m = 3  # Количество видов ценных бумаг
n = [100, 200, 150]  # Количество ценных бумаг каждого вида
S_0 = [100, 50, 80]  # Начальная стоимость ценных бумаг каждого вида
mu = [0.05, 0.03, 0.04]  # Ожидаемая доходность
sigma = [0.2, 0.15, 0.25]  # Волатильность
T = 1  # Временной горизонт

# Начальная стоимость портфеля
V_0 = sum(n[i] * S_0[i] for i in range(m))

N = 1000  # Количество симуляций Монте-Карло
# Генерация случайных величин Z_i (стандартное нормальное распределение)
Z = np.random.normal(0, 1, (N, m))

# Моделирование цен активов для всех сценариев
S_t = np.zeros((N, m))
for i in range(m):
    S_t[:, i] = S_0[i] * np.exp((mu[i] - 0.5 * sigma[i]**2) * T + sigma[i] * np.sqrt(T) * Z[:, i])

# Стоимость портфеля для каждого сценария
V_t = np.dot(S_t, n)

# Убытки
L_t = V_0 - V_t
# Расчет VaR
# с вероятностью 95% потери не превысят VaR
alpha = 0.05
VaR = np.percentile(L_t, 100 * (1 - alpha))

print(f"VaR на уровне {1-alpha:.0%}: {VaR:.2f}")
```
